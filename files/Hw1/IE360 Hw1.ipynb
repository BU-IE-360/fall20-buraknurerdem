# Measure 1: Housing and Rent

Here are the histograms of housing and rent expenditure percantages of households in different regions by year.

```r
houserent = read.csv("HousingRent.csv")
rownames(houserent) = houserent[1:26,1] #Naming Row Names
houserent = houserent[,-1]              #Dropping first column which was the row names
colnames(houserent) = c(2003:2017)      #Also changed to column names because it showed "X2003", "X2004"...

houserent

par(mfrow=c(4,4))
for (i in 1:length(houserent[1,]))
    hist(houserent[,i],
         main = paste("Year",colnames(houserent)[i]) ,
         xlab = "Spending% in Housing",
         ylab = "Frequency of Regions",
         xlim = c(15,35))
```

In the histograms, it looks like there are not a drastic change over the years except that spending has decreased over the last few years. Maybe, we can see the changes better by plotting regions individually.

```r
colVector = c("#78b98f", "#0b522e", "#bae342", "#a1173d", "#1cf1a3",
            "#2c457d", "#e9c9fa", "#791c76", "#f597fa", "#35c8ef",
            "#888bce", "#4224bf", "#46f33e", "#db11ac", "#3aa609",
            "#c47f6f", "#693a30", "#f8cca6", "#ec4b18", "#f4d403",
            "#f35c79", "#d6790b", "#9f04fc", "#0e8f6d", "#452e48",
            "#69cec3", "#85cd5d", "#3c0958", "#074a68", "#7ccad5",
            "#b9f78f") # I tried to find 26 distinctive colors.

plot(x = colnames(houserent),
     y = houserent[1,],
     type = "o",
     xlab = "Years", ylab = "Percentage",     
     lwd = 2, col = colVector[1], pch=0,    
     ylim = c(15, 36))

for(i in 2:26)
    lines(x = colnames(houserent),
          y = houserent[i,],
          col = colVector[i], pch = i-1,          
          lwd = 2, type = "o")
title("Housing and Rent Expenditure")
legend("topright", rownames(houserent), cex=0.7, pt.cex=1.2,
       pch=c(0:25),col=colVector, text.col=colVector, lwd=2, ncol= 4, inset = c(0.01, 0.01))

abline(v=(seq(2003,2017,1)), col="lightgray", lty="dotted")
abline(h=(seq(15,35,1)), col="lightgray", lty="dotted")
```

It is possible to say that we can see a slightly decreasing trend in this graph. Conversely, there some regions where the percantages have increased over time. In another aspect, it is clearly seen that the lines get cramped over time, therefore, we can say variation of percantage spending in housing and rent in different regions have become smaller.

Below, we will inspect the data of search results "kiralık ev" and "satılık ev" from Google Trends for Turkey. However, be noted that year 2020 is not finished at the moment, values of 2020 will not be accurate. And Google Trends data start with year 2004, not 2003. Similarly, the expenditure data ends in 2017.

```r
rentalHouse = read.csv("kiralikev.csv")
forsaleHouse = read.csv("satilikev.csv")

# We need to put these monthly data in yearly format

rentalHouseYearly = seq(0,0, length.out = length(rentalHouse[,1])/12 )
forsaleHouseYearly = seq(0,0, length.out = length(forsaleHouse[,1])/12 )

dummyVec = rep(1:ceiling(length(rentalHouse[,1])/12), length.out = length(rentalHouse[,1]), each = 12)

for(i in 1:length(rentalHouse[,2])){
    rentalHouseYearly[dummyVec[i]] = rentalHouseYearly[dummyVec[i]] + rentalHouse[i,2]
    }
rentalHouseYearly = rentalHouseYearly / 12

for(i in 1:length(forsaleHouse[,2])){
    forsaleHouseYearly[dummyVec[i]] = forsaleHouseYearly[dummyVec[i]] + forsaleHouse[i,2]
    }
forsaleHouseYearly = forsaleHouseYearly / 12
    
# I know that this is not the most conventional method. I need to find a better way to do this.
# But I checked the values and they are correct. This will do for now.

# Let's calculate means of years of housing and renting expeniture percentage.
# Below, I took the arithmetic averages of regions, 
# however, it would be more accurate to take population weighted averages.

meansOfHouseRents = seq(0, 0, length.out = length(houserent[1,]))
for(i in 1:length(houserent[1,])){
    meansOfHouseRents[i] = mean(houserent[,i])
    }

# Let's plot

plot(x = c(2004:2020), y = rentalHouseYearly,
     col = "red", lwd = 2, pch = 1, type = "o",
     xlab = "Years", ylab = "Percentage", xlim = c(2003,2020), ylim = c(0,100),
     main = "Google Trends Results")
lines(x = c(2004:2020), y = forsaleHouseYearly,
      col="blue", lwd=2, pch=2, type="o") 
lines(x = c(2003:2017), y = meansOfHouseRents, 
      col="orange", lwd=2, pch=4, type="o")
legend("topright", c("Rental House Search", "For Sale House Search", "Housing Expenditure"), 
       col = c("red", "blue", "orange"), text.col = c("red", "blue", "orange"),
       pch = c(1,2,4), lwd = 2, inset = c(0.01, 0.01), cex = 1.2, pt.cex = 2)       

abline(v=(seq(2003,2020,1)), col="lightgray", lty="dotted")
abline(h=(seq(0,100,10)), col="lightgray", lty="dotted")
```

Now, before commenting on the graph, it should be noted that the trend percentages' means and the spending percentages differ in nature. What I mean is that the ways that those two datasets calculated probably doesn't allow them to be compared directly. It is arguable whether they should be plotted on the same graph. I plotted anyway for visual purposes.

Google search results of rental housing and for sale housing show increasing trends, while expenditure percentage of housing and renting are fairly constant over time. My opinion on this subject is that since housing is such a basic need, for its percantage to change drastically over time there has to be a national or global economic boom or crisis. It is not suprising that it is fairly constant. However, increasing trends of search results show us, in my opinion, that the medium of looking for houses to buy or rent has moved to internet over time with the help of widespreading usage of technology and internet. It looks like people started to prefer the internet to search for suitable housing compared to old methods such as touring neighborhoods and calling numbers on the windows or consulting real estate offices.
